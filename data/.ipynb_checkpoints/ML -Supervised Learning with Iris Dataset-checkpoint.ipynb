```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import model_selection
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Imputer
import pickle
from sklearn.externals import joblib

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC



# load our dataset
df = pd.read_csv("heart.csv")

df.head()

df.describe()

df.shape

#Check Null Value
print(df.isnull().values.any())

#Correlation Heat Map
def corr_heatmap(df, size=11):
  # Getting correlation using Pandas
  correlation = df.corr()

  # Dividing the plot into subplots for increasing size of plots
  fig, heatmap = plt.subplots(figsize=(size, size))

  # Plotting the correlation heatmap
  heatmap.matshow(correlation)

  # Adding xticks and yticks
  plt.xticks(range(len(correlation.columns)), correlation.columns)
  plt.yticks(range(len(correlation.columns)), correlation.columns)

  # Displaying the graph
  plt.show()

corr_heatmap(df, size=30)

#True False Ratio Check
num_true = 0.0
num_false = 0.0
for item in df['target']:
    if item == 1:
        num_true += 1
    else:
        num_false += 1

percent_true = (num_true / (num_true + num_false)) * 100
percent_false = (num_false / (num_true + num_false)) * 100

print("Number of True Cases:",percent_true)
print("Number of False Cases:",percent_false)

feature_column_names = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']
predicted_class_name = ['target']

# Getting feature variable values
X = df[feature_column_names].values
y = df[predicted_class_name].values

# Saving 30% for testing
split_test_size = 30

# Splitting using scikit-learn train_test_split function
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = split_test_size, random_state = 42)

print("X_Train Shape",X_train.shape)
print("y_train Shape",y_train.shape)

print("X_test Shape",X_test.shape)
print("y_test Shape",y_test.shape)

#Check Missing Value
print("# rows in dataframe {0}".format(len(df)))
print("# rows missing age: {0}".format(len(df.loc[df['age'] == 0])))
print("# rows missing cp: {0}".format(len(df.loc[df['cp'] == 0])))
print("# rows missing trestbps: {0}".format(len(df.loc[df['trestbps'] == 0])))
print("# rows missing chol: {0}".format(len(df.loc[df['chol'] == 0])))
print("# rows missing fbs: {0}".format(len(df.loc[df['fbs'] == 0])))
print("# rows missing restecg: {0}".format(len(df.loc[df['restecg'] == 0])))
print("# rows missing thalach: {0}".format(len(df.loc[df['thalach'] == 0])))
print("# rows missing exang: {0}".format(len(df.loc[df['exang'] == 0])))
print("# rows missing oldpeak: {0}".format(len(df.loc[df['oldpeak'] == 0])))
print("# rows missing slope: {0}".format(len(df.loc[df['slope'] == 0])))
print("# rows missing ca: {0}".format(len(df.loc[df['ca'] == 0])))
print("# rows missing thal: {0}".format(len(df.loc[df['thal'] == 0])))

#Impute with mean all 0 readings
fill_0 = Imputer(missing_values=0, strategy="mean", axis=0)

X_train = fill_0.fit_transform(X_train)
X_test = fill_0.fit_transform(X_test)


######################### Logestic Regression #########################
lr_model = LogisticRegression(C=0.7, random_state=42)
lr_model.fit(X_train, y_train.ravel())
lr_predict_test = lr_model.predict(X_test)

print("Acuuracy", metrics.accuracy_score(y_test, lr_predict_test))
print("Confusion Matrix")
print(metrics.confusion_matrix(y_test, lr_predict_test, labels=[1, 0]))
cm = metrics.confusion_matrix(y_test, lr_predict_test, labels=[1, 0])
print("Classification Report For Logestic Regression")
print(metrics.classification_report(y_test, lr_predict_test, labels=[1, 0]))

mysample = np.array([[62,1,2,130,203,0,1,152,0,1,3,2,1]])
prediction = lr_model.predict(mysample)
print("prediction : " , prediction)

# save the model to disk
filename = 'logit_model_heart.pkl'
pickle.dump(lr_model, open(filename, 'wb'))


# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, y_test)
print(result)

mysample = np.array([[62,1,2,130,203,0,1,152,0,1,3,2,1]])
prediction = loaded_model.predict(mysample)
print("prediction : " , prediction)

# save the model to disk
joblib.dump(lr_model, 'logit_model_heart2.pkl')

# Reloading the Model
loaded_model = joblib.load('logit_model_heart2.pkl')
result = loaded_model.score(X_test, y_test)
print(result)

mysample = np.array([[62,1,2,130,203,0,1,152,0,1,3,2,1]])
prediction = loaded_model.predict(mysample)
print("prediction : " , prediction)
######################### Logestic Regression #########################

#######################################  Gaussian Naive Bayes ##########################
# create Gaussian Naive Bayes model object and train it with the data
nb_model = GaussianNB()
nb_model.fit(X_train, y_train.ravel())

# this returns array of predicted results from test_data
prediction_from_test_data = nb_model.predict(X_test)
accuracy2 = metrics.accuracy_score(y_test, prediction_from_test_data)

print("Test Data Accuracy : " , accuracy2)
print("Confusion Matrix")
print(metrics.confusion_matrix(y_test, prediction_from_test_data, labels=[1, 0]))
print("Classification Report Naive Bayes model")
print(metrics.classification_report(y_test, prediction_from_test_data, labels=[1, 0]))

# save the model to disk
joblib.dump(nb_model, 'nb_model_heart.pkl')
#####################################################################################################

######################################### RandomForestClassifier #################
# Create a RandomForestClassifier object
rf_model = RandomForestClassifier(random_state=42)

rf_model.fit(X_train, y_train.ravel())
rf_predict_test = rf_model.predict(X_test)

#get accuracy
rf_accuracy_testdata = metrics.accuracy_score(y_test, rf_predict_test)

#print accuracy
print("Accuracy:" , rf_accuracy_testdata)
print("Confusion Matrix for Random Forest")
print(metrics.confusion_matrix(y_test, rf_predict_test, labels=[1, 0]))
print(metrics.classification_report(y_test, rf_predict_test, labels=[1, 0]))

# save the model to disk
joblib.dump(rf_model, 'rr_model_heart.pkl')

##########################################################

###################### KNeighborsClassifier ##############
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train , y_train.ravel())
knn_predict_test = knn.predict(X_test)
knn_accuracy_testdata = knn.score(X_test , y_test)

print("Accuracy: " , knn_accuracy_testdata)
print("Confusion Matrix for KNeighborsClassifier")
print(metrics.confusion_matrix(y_test, knn_predict_test, labels=[1, 0]))
print(metrics.classification_report(y_test, knn_predict_test, labels=[1, 0]))

# save the model to disk
joblib.dump(knn, 'knn_model_heart.pkl')
##################################################################

###################### DecisionTreeClassifier() ##############
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train.ravel())
dtree_predict_test = dtree.predict(X_test)
dtree_accuracy_testdata = dtree.score(X_test , y_test)

print("Accuracy: " , dtree_accuracy_testdata)
print("Confusion Matrix for DecisionTreeClassifier")
print(metrics.confusion_matrix(y_test, dtree_predict_test, labels=[1, 0]))
print(metrics.classification_report(y_test, dtree_predict_test, labels=[1, 0]))

# save the model to disk
joblib.dump(dtree, 'dtree_model_heart.pkl')
##################################################################

######################### SVM #########################
svm = SVC()
svm.fit(X_train, y_train.ravel())
svm_predict_test = svm.predict(X_test)
svm_accuracy_testdata = svm.score(X_test , y_test)

mysample = np.array([[62,1,2,130,203,0,1,152,0,1,3,2,1]])
prediction = svm.predict(mysample)
print("prediction : " , prediction)

print("Accuracy: " , svm_accuracy_testdata)
print("Confusion Matrix for SVM")
print(metrics.confusion_matrix(y_test, svm_predict_test, labels=[1, 0]))
print(metrics.classification_report(y_test, svm_predict_test, labels=[1, 0]))


# save the model to disk
joblib.dump(svm, 'svm_model_heart.pkl')
######################### SVM #########################
```
